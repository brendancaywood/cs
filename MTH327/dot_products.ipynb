# Lengths and Dot Products

Remember that you may define a vector, such as ${\mathbf v} = \left[\begin{array}{c}-2 \\ 3 \\ 1 \\ \end{array}\right]$, as

<code>v = vector([-2, 3, 1])</code>

The length of the vector is <code> v.norm() </code> and the dot product of two vectors <code>v</code> and <code>w</code>code> is <code>v.dot_product(w)</code>.

Define two vectors $v=\left[\begin{array}{c}2 \\ 0 \\ 3 \\ -2 \\ \end{array}\right]$ and 
$w=\left[\begin{array}{c}1 \\ -3 \\ 4 \\ 1 \end{array}\right]$ in the cell below.

```python
v = vector([2, 0, 3, -2]) 
w = vector([1, -3, 4, 1])
v.dot_product(w)
```

Find the lengths $|{\mathbf v}|$, $|{\mathbf w}|$ and the dot product ${\mathbf v}\cdot{\mathbf w}$.  You may use <code>numerical_approx(a)</code> to see a numerical approximation of a quantity <code>a</code> rather than its exact value.

```python
v = vector([2, 0, 3, -2]) 
w = vector([1, -3, 4, 1])
a =  numerical_approx(v.norm())
b = numerical_approx(w.norm())
print numerical_approx(arccos(12 / (a * b)) * (180 / pi))
```

## Word counts in Wikipedia articles

Evaluating the cell below will create five vectors, called <code>veterans</code>, <code>memorial</code>, <code>labor</code>, <code>golden</code>, and <code>super</code>.   

```python
import csv
reader = csv.reader(open("events.csv"))
next(reader)
data = [row for row in reader]
freqs = zip(*data)
freqs.pop(0)
freqs = [map(float, col) for col in freqs]
freqs = [vector(col) for col in freqs]
veterans, memorial, labor, golden, super = freqs
print "The vector 'veterans' is " + str(len(veterans)) + "-dimensional"
```

To create these vectors, I pulled the text from the wikipedia articles, removed common words like "a" and "the", and looked at words that appeared in at least two of the articles.  For instance, evaluate the next cell to look at the vector <code>memorial</code>.

## Time series and correlations

A vector may be used to represent a *time series*, which we will now explain.  Evaluate the following cells to load in two vectors called <code>s1</code> and <code>s2</code>.  You will also see a plot of the components of the vectors.

```python
import csv
reader = csv.reader(open("time-series.csv"))
data = [map(float, row) for row in reader]
data = map(vector, zip(*data))
s1 = data[0]
s2 = data[3]
s3 = data[1]

def seriesplot(series, color):
    return list_plot(series, color=color, plotjoined=True) + list_plot(series, color=color, size=20)

seriesplot(s1, 'blue') + seriesplot(s2, 'green')
```

This vector could represent, say, the value of a stock over time.  So for <code>s1</code>, the vector represented in blue, the value on the first day is about 4.2, the second day is about 7.5, and so forth.  The two vectors could represent the time series of two different companies.  Notice that the two time series seem to be related;  when one rises, the other seems to rise too.  This would be important information to know if you are an investor, and we can measure it with a quantity called *correlation*.

The first thing to note is that the blue values are generally higher than the green ones, just as the value of Google would be higher than, say, Fishback Enterprises.  The first step is to look at the average values of the stocks and subtract them.  This is called *de-meaning* the time series, and it can be accomplished by evaluating the next cell.  The new de-meaned vectors are <code>ds1</code> and <code>ds2</code>.

```python
import numpy as np
def demean(series):
    return vector(series - mean(series)*np.ones(len(series)))

ds1 = demean(s1)
ds2 = demean(s2)
seriesplot(ds1, 'blue') + seriesplot(ds2, 'green')
```

The correlation between two vectors ${\mathbf v}$ and $\mathbf w$ is defined to be 
$$
\mbox{corr}({\mathbf v}, {\mathbf w}) = \frac{{\mathbf v}\cdot{\mathbf w}}{|{\mathbf v}|~|{\mathbf w}|}
$$
and the correlation between the two time series is the correlation between their de-meaned vectors.  

Explain why the correlation is always between -1 and 1.  Then find the correlation between the time series <code>s1</code> and <code>s2</code> (remember to use their de-meaned vectors).
